```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("testset.csv",parse_dates=['datetime_utc'],skipinitialspace=True)
```

```python
data['Date'] = pd.to_datetime(data['datetime_utc'])
data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month
data['day'] = data['Date'].dt.day
data['hour'] = data['Date'].dt.hour
```

# 1. Cleaning of data

```python
data.fillna(0, inplace=False)
```

```python
data=data.drop('_precipm',axis=1)
```

```python
data.head()
```

# 2. linear plot

```python
year_humi = data.groupby(data.year).mean()
fig,axes = plt.subplots(nrows=4,ncols=4,figsize=(25,40))
fig.subplots_adjust(hspace=0.5)
year_humi['_hum'].plot(ax = axes[0,0])
axes[0,0].set_title("humidity")
year_humi['_rain'].plot(ax = axes[0,1])
axes[0,1].set_title("Rain")
year_humi['_dewptm'].plot(ax = axes[0,2])
axes[0,2].set_title("Dew ")
year_humi['_pressurem'].plot(ax = axes[0,3])
axes[0,3].set_title("Pressure")
year_humi['_fog'].plot(ax = axes[1,0])
axes[1,0].set_title("Fog")
year_humi['_hail'].plot(ax = axes[1,1])
axes[1,1].set_title("Hail")
year_humi['_heatindexm'].plot(ax = axes[1,2])
axes[1,2].set_title("Heat index")
year_humi['_snow'].plot(ax = axes[1,3])
axes[1,3].set_title("Snow")
year_humi['_tornado'].plot(ax = axes[2,0])
axes[2,0].set_title("Tornado")
year_humi['_vism'].plot(ax = axes[2,1])
axes[2,1].set_title("Vism")
year_humi['_wdird'].plot(ax = axes[2,2])
axes[2,2].set_title("Wind Direction")
year_humi['_wspdm'].plot(ax = axes[2,3])
axes[2,3].set_title("Wind Speed")
year_humi['_tempm'].plot(ax = axes[3,0])
axes[3,0].set_title("Temprature")
year_humi['_thunder'].plot(ax = axes[3,1])
axes[3,1].set_title("Thunder")
```

# 3. Pie chart

```python
act=["Haze",'Fog','Smoke','Mist','Shallow Fog']
colors=['r','g','b','grey','magenta']
slices=[]
for i in act:
    slices.append(len(data[(data['month']==1) & (data['_conds']==i) ]))
    
plt.pie(slices, labels=act,colors=colors,shadow=True,autopct="%1.1f%%",startangle=90)
plt.legend(loc=(1,1))
plt.title("january")
plt.show()
```
