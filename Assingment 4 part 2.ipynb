```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("testset.csv",parse_dates=['datetime_utc'],skipinitialspace=True)
```

```python
data['Date'] = pd.to_datetime(data['datetime_utc'])
data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month
data['day'] = data['Date'].dt.day
data['hour'] = data['Date'].dt.hour
```

# 1. Cleaning of data

```python
data.fillna(0, inplace=False)
```

```python
data=data.drop('_precipm',axis=1)
```

```python
data.head()
```

# 2. linear plot

```python
year_humi = data.groupby(data.year).mean()
fig,axes = plt.subplots(nrows=1,ncols=4,figsize=(15,5))
fig.subplots_adjust(hspace=0.8)
year_humi['_hum'].plot(ax = axes[0])
axes[0].set_title("humidity")
year_humi['_rain'].plot(ax = axes[1])
axes[1].set_title("Rain")
year_humi['_dewptm'].plot(ax = axes[2])
axes[2].set_title("Dew ")
year_humi['_pressurem'].plot(ax = axes[3])
axes[3].set_title("Pressure")
fig,axe = plt.subplots(nrows=1,ncols=4,figsize=(15,5))
year_humi['_fog'].plot(ax = axe[0])
axe[0].set_title("Fog")
year_humi['_hail'].plot(ax = axe[1])
axe[1].set_title("Hail")
year_humi['_heatindexm'].plot(ax = axe[2])
axe[2].set_title("Heat index")
year_humi['_snow'].plot(ax = axe[3])
axe[3].set_title("Snow")
fig,ax = plt.subplots(nrows=1,ncols=4,figsize=(15,5))
year_humi['_tornado'].plot(ax = ax[0])
ax[0].set_title("Tornado")
year_humi['_vism'].plot(ax = ax[1])
ax[1].set_title("Vism")
year_humi['_wdird'].plot(ax = ax[2])
ax[2].set_title("Wind Direction")
year_humi['_wspdm'].plot(ax = ax[3])
ax[3].set_title("Wind Speed")
fig,a = plt.subplots(nrows=1,ncols=2,figsize=(8,5))
year_humi['_tempm'].plot(ax = a[0])
a[0].set_title("Temprature")
year_humi['_thunder'].plot(ax = a[1])
a[1].set_title("Thunder")


#plt.title('Delhi Average Humidity by year')
#plt.xlabel('year')
```

# 3. Pie chart

```python
act=["Haze",'Fog','Smoke','Mist','Shallow Fog']
colors=['r','g','b','grey','magenta']
slices=[]
for i in act:
    slices.append(len(data[(data['month']==1) & (data['_conds']==i) ]))
    
plt.pie(slices, labels=act,colors=colors,shadow=True,autopct="%1.1f%%",startangle=90)
plt.legend(loc=(1,1))
plt.title("january")
plt.show()
```
